# Probabilistic Agent Simulations and Stats

```python
import os
import sys
import statistics as st
from matplotlib import pyplot as plt

os.chdir("..")

from wumpusWorld.agent.ProbAgent import *
from wumpusWorld.agent.ConditionalProbs import *
from wumpusWorld.environment.Environment import *
from wumpusWorld.environment.Percept import *
from wumpusWorld.environment.Agent import *
from wumpusWorld.environment.Action import *
```

```python
def runScenario():
    environment = Environment.create(pitProb = 0.2)
    percept = Percept.create(environment.isStench(Coords(0,0)), 
                             environment.isBreeze(Coords(0,0)), 
                             environment.isGlitter(Coords(0,0)))
    pitModel = createPitModel(environment)
    wumpusModel = createWumpusModel(environment)
    probAgent = ProbAgent(pitModel, wumpusModel)
    agent = ProbAgent.create()
    while not environment.terminated:
        environment.visualize(agent)
        percept.show()
        agent.show()
        action = probAgent.nextAction(percept, agent)
        percept, agent = environment.applyAction(percept, action, agent)
    percept.show()
    agent.show()
    return percept, agent
```

## Part 2: 1000 Runs

We need HiddenPrints class to make sure the code doesn't print the simulations for 1000 runs

```python
class HiddenPrints:
    def __enter__(self):
        self._original_stdout = sys.stdout
        sys.stdout = open(os.devnull, 'w')

    def __exit__(self, exc_type, exc_val, exc_tb):
        sys.stdout.close()
        sys.stdout = self._original_stdout
```

```python
rewardList = []
gameCnt = 1000
deathCnt = 0
shotCnt = 0

with HiddenPrints():
    for i in range(gameCnt):
        perceptVar, agentVar = runScenario()
        rewardList.append(perceptVar.reward)
        if agentVar.isAlive == False: deathCnt+=1
        if agentVar.hasArrow == False: shotCnt+=1
```

```python
print('Total Reward after 1000 Games: ' + str(sum(rewardList)))
print('Average Reward after 1000 Games: ' + str(sum(rewardList)/1000))
print('Median Reward after 1000 Games: ' + str(st.median(rewardList)))
print('Number of time the agent won: ' + str(len([i for i in rewardList if i>0])))
print('Number of times the agent died: ' + str(deathCnt))
print('Number of times the agent shot the arrow: ' + str(shotCnt))
```

```python
plt.hist(rewardList)
plt.title('Distribution of Rewards')
plt.xlabel('Reward')
plt.ylabel('Count')

plt.show()
```
